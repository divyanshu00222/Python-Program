```python
#✅ 1️⃣ Find factorial using recursion 

def fact(n):
    return 1 if n==0 else n * fact(n-1) 
print(fact(5)) 
```

```python
#✅ 2️⃣ Find common elements in two lists 

a = [1,2,3,4]
b = [3,4,5]
print(list(set(a) & set(b)))
```

```python
#✅ 3️⃣ Count vowels using lambda + filter 

s = "developer" 
vowels = list(filter(lambda x: x in "aeiouAEIOU", s))
print(len(vowels)) 
```

```python
#✅ 4️⃣ Check if list is palindrome 

arr = [1,2,3,2,1]
print(arr == arr[::-1]) 
```

```python
#✅ 5️⃣ Find largest word in sentence 

s = "Python makes coding simple" 
print(max(s.split(), key= len)) 
```
