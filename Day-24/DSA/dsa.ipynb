```python
#Q1. Reverse a string using Stack 

def reverse_string(s):
    stack = []
    for ch in s:
        stack.append(ch)
    rev = ""
    while stack:
        rev += stack.pop()
    return rev

print(reverse_string("divyanshu"))
```

```python
#Q2. Check Balanced Parentheses (VERY IMPORTANT)

# () -> valid
# (] -> invalid

def is_balanced(expr):
    stack = []
    pairs = {')':'(', ']':'[', '}':'{'}

    for ch in expr:
        if ch in "([{":
            stack.append(ch)
        elif ch in ")]}":
            if not stack or stack.pop() != pairs[ch]:
                return False
    return len(stack)==0

print(is_balanced("{[()]}"))
```

```python
#Q3. Implement Queue Using 2 Stacks 

class MyQueue:
    def __init__(self):
        self.s1 = []
        self.s2 = []

    def enqueue(self, x):
        self.s1.append(x)

    def dequeue(self):
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2.pop()
```

```python
#Q4. Find next greater element 

def next_greater(arr):
    stack = []
    res = [-1] * len(arr)

    for i in range(len(arr)-1, -1, -1):
        while stack and stack[-1] <= arr[i]:
            stack.pop()

        if stack:
            res[i] = stack[-1]

        stack.append(arr[i])

    return res

print(next_greater([4,5,2,10,8]))
```

```python
#Q5. Queue using List 

class Queue:
    def __init__(self):
        self.q = []

    def enqueue(self, x):
        self.q.append(x)

    def dequeue(self):
        if not self.q:
            return None
        return self.q.pop(0)
```
