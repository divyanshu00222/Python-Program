Real life example ‚Üí Undo, Browser Back, Call Stack

```python
#‚úî Stack using Python List 

stack = [] 

stack.append(10)   # push
stack.append(20)
stack.append(30)

print(stack.pop())   # pop - 30
print(stack[-1])     # peek - 20
```

```python
#‚úî Stack using Collections.deque (FASTER) 

from collections import deque

stack = deque()

stack.append(10)
stack.append(20)
stack.append(30)

print(stack.pop())     # 30
print(stack[-1])       # 20

```
