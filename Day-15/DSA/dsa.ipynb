```python
#✅ 1️⃣ Find sum of array recursively 

def sum_arr(arr):
    if not arr:
        return 0 
    return arr[0] + sum_arr(arr[1:])

print(sum_arr([1,2,3,4,5])) 
```

```python
#✅ 2️⃣ Find height of a binary tree 

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def height(root):
    if not root:
        return 0
    return 1 + max(height(root.left), height(root.right))

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
print("Height:", height(root))
```

```python
#✅ 3️⃣ Count leaf nodes 

def count_leaves(root):
    if not root:
        return 0
    if not root.left and not root.right:
        return 1
    return count_leaves(root.left) + count_leaves(root.right)
```

```python
#✅ 4️⃣ Find maximum element in BST 

def max_in_bst(root):
    while root.right:
        root = root.right
    return root.key
```

```python
# ✅ 5️⃣ Reverse array recursively

def reverse(arr):
    if len(arr) <= 1:
        return arr
    return [arr[-1]] + reverse(arr[:-1])

print(reverse([1,2,3,4,5]))
```
