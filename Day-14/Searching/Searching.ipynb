```python
#✅ Linear Search 

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return -i 
    return -1


arr = [10,20,30,40,50]
print(linear_search(arr,30))
```

```python
#✅ Binary Search 

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [10, 20, 30, 40, 50]
print(binary_search(arr, 40))  # Output: 3

# ✅ Used in: Databases, search engines, sorted arrays
```
