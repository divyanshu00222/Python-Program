```python
#✅ 1️⃣ Find smallest and largest in array 

arr = [10,20,4,45,99]
print(min(arr),max(arr)) 
```

```python
#✅ 2️⃣ Find second largest number 

arr = [10, 20, 4, 45, 99]
arr = list(set(arr))
arr.sort()
print(arr[-2])
```

```python
#✅ 3️⃣ Count number of occurrences 

arr = [1, 2, 2, 3, 3, 3]
count = {x: arr.count(x) for x in set(arr)}
print(count)
```

```python
#✅ 4️⃣ Find missing element 

arr = [1, 2, 3, 5]
n = 5
print(n*(n+1)//2 - sum(arr))
```

```python
#✅ 5️⃣ Merge two sorted arrays 

a = [1, 3, 5]
b = [2, 4, 6]
print(sorted(a + b))
```
