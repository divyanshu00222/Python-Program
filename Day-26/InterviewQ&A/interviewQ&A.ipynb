```python
#Q1. Reverse words in a sentence 
def reverse_words(s):
    return " ".join(s.split()[::-1])

```

```python
#Q2. Find duplicates in list 

def find_duplicates(arr):
    seen = set()
    dup = set()
    for x in arr:
        if x in seen: dup.add(x)
        seen.add(x)
    return list(dup)
```

```python
#Q3. Count vowels 

def count_vowels(s):
    return sum(1 for ch in s.lower() if ch in "aeiou")
```

```python
#Q4. Check anagram 

def is_anagram(a,b):
    return sorted(a) == sorted(b)
```

```python
#Q5. Find missing number 

def missing(arr):
    n = len(arr) + 1
    return n*(n+1)//2 - sum(arr)
```
