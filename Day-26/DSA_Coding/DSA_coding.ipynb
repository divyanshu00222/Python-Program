```python
#1. Reverse Linked List

def reverse_ll(head):
    prev = None
    cur = head
    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur
        cur = nxt
    return prev
```

```python
#2. Check if Linked List has cycle 

def has_cycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
```

```python
#4. First non-repeating character 

from collections import Counter

def first_unique(s):
    c = Counter(s)
    for ch in s:
        if c[ch] == 1:
            return ch
    return -1
```

```python
#5. Majority Element 

def majority(arr):
    count = 0
    candidate = None

    for num in arr:
        if count == 0:
            candidate = num
        count += 1 if num == candidate else -1

    return candidate
```

```python
#6. Merge two sorted arrays 

def merge(a, b):
    return sorted(a + b)
```

```python
#7. Rotate array 
def rotate(arr, k):
    k %= len(arr)
    return arr[-k:] + arr[:-k]

```

```python
#8. Max subarray sum (Kadane) 

def max_sub(arr):
    best = cur = arr[0]
    for x in arr[1:]:
        cur = max(x, cur + x)
        best = max(best, cur)
    return best
```

```python
#9. Second largest 

def second_largest(arr):
    arr = sorted(set(arr))
    return arr[-2]
```

```python
#10. Count digits 

def count_digits(n):
    return len(str(n))
```
