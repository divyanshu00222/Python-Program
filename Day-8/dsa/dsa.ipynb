```python
#✅ 1️⃣ Find first non-repeating character 

s = "aabbcdde" 
for ch in s:
    if s.count(ch) == 1:
        print(ch) 
        break
```

```python
#✅ 2️⃣ Move all zeros to end 

arr = [0,1,0,3,12]
result = [x for x in arr if x!=0] + [0]*arr.count(0) 
print(result) 
```

```python
#✅ 3️⃣ Intersection of two arrays 

a = [1,2,2,3]
b = [2,2]
print(list(set(a) & set(b))) 
```

```python
#✅ 4️⃣ Check if two strings are anagrams 

s1 = "listen" 
s2 = "silent" 
print(sorted(s1) == sorted(s2)) 
```

```python
#✅ 5️⃣ Find all pairs with sum = target 

arr = [1,3,5,7,9]
target = 10 

for i in range(len(arr)):
    for j in range(i+1, len(arr)):
        if arr[i] + arr[j] == target:
            print(arr[i], arr[j]) 
```
