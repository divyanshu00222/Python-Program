Q: Difference between list, tuple, set, dictionary?

✅ Answer:

Type	Ordered	      Mutable	        Duplicate Allowed	        Syntax
List	✅ Yes	      ✅ Yes	        ✅ Yes	                   []
Tuple	✅ Yes	      ❌ No	        ✅ Yes	                   ()
Set	    ❌ No	      ✅ Yes	        ❌ No	                   {}
Dict	✅ Yes (keys)  ✅ Yes	    Keys ❌ dup	               {"key": value}
