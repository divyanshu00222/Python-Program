```python
# ✅ 1. Lists

# Ordered, mutable.

numbers = [10,20,30]
numbers.append(40) # add
numbers.remove(20) # delete
print(numbers[1]) # access
```

```python
# ✅ 2. Tuples

# Ordered, immutable.

t = (1,2,3)
print(t[0])
```

```python
# ✅ 3. Sets

# Unordered, no duplicates.

s = {1,2,2,3}
print(s) # {1,2,3}

s.add(5)
s.remove(3)
```

```python
# ✅ 4. Dictionaries

# Key-value pairs.

student = {
    "name": "Divyanshu",
    "age" : 23,
    "course" : "Python"
}
print(student["name"])
student["age"] = 23 #update
```

```python
# ✅ 5. List Comprehension (VERY IMPORTANT)

squares = [x*x for x in range(1, 6)]
print(squares)
```
