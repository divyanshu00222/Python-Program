```python
# 1️⃣ Create a list & print sum, max, min

nums = [10,25,5,40,7]
print("Sum =", sum(nums))
print("Max =", max(nums))
print("min =", min(nums))
```

```python
# 2️⃣ Remove duplicates (list → set → list)

nums = [1,2,2,3,3,4]
unique = list(set(nums))
print(unique)
```

```python
# 3️⃣ Create dictionary of student info

student = {
    "name" : "Divyanshu",
    "age" : 22,
    "course" : "Python"
}

for key , value in student.items():
    print(key, ":", value)
```

```python
# 4️⃣ Count frequency of elements using dictionary

nums = {1,2,2,3,3,3}
freq = {}
for n in nums :
    freq[n] = freq.get(n , 0) + 1

print(freq)
```

```python
# 5️⃣ List comprehension: create list of squares 1–10
squares = [x*x for x in range(1, 11)]
print(squares)
```
