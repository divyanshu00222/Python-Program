```python
# 1️⃣ Find the element that appears only once

# (arr where others appear twice)

arr = [2,3,2,4,4]
unique = 0 
for num in arr:
    unique ^= num 
print(unique)
```

```python
# 2️⃣ Check if array is sorted

is_sort = True
for i in range(len(arr)-1):
    if arr[i] > arr[i+1]:
        is_sort = False
        break
print(is_sort)
```

```python
# 3️⃣ Move all zeros to end

arr = [0,1,0,3,12]
result = [x for x in arr if x != 0] + [0]*arr.count(0)
print(result)
```

```python
# 4️⃣ Find intersection of two arrays

a = [1,2,3,4]
b = [3,4,5]

print(list(set(a) & set(b)))
```

```python
# 5️⃣ Reverse words in a sentence

sentence = "hello world python" 
words = sentence.split()
print(" ".join(words[::-1]))
```
