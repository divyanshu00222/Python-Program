```python
#✅ 2️⃣ Implement queue using two stacks 

s1, s2 = [],[]

def enqueue(x):
    s1.append(x) 

def dequeue():
    if not s2:
        while s1:
            s2.append(s1.pop())
        return s2.pop() if s2 else None
    
enqueue(1)
enqueue(2)
enqueue(3)
print(dequeue()) #1
```
