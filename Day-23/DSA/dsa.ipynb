```python
#3️⃣ Find second largest number 

def second_largest(arr):
    arr = list(set(arr))
    arr.sort()
    return arr[-2]
```

```python
#4️⃣ Count how many numbers are greater than X 

def count_greater(arr, x):
    return sum(1 for i in arr if i > x)
```

```python
#5️⃣ Return True if array is sorted 

def is_sorted(arr):
    return all(arr[i] <= arr[i+1] for i in range(len(arr)-1))
```
