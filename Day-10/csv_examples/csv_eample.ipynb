```python
#✅ Write into CSV 
import csv 
with open("Student.csv","w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Roll","Name","Marks"])
    writer.writerow([1,"Divyanshu",90])
```

```python
#✅ Read CSV 
import csv 

with open("Student.csv","r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row) 
```
