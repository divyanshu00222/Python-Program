```python
#✅ 1️⃣ Find all unique elements 

arr = [1,2,2,3,4,4,5]
unique = list(set(arr))
print(unique) 
```

```python
#✅ 2️⃣ Reverse list without using reverse() 

arr = [1,2,3,4]
print(arr[::-1]) 
```

```python
#✅ 3️⃣ Check if a string has all unique characters 

s = "Python" 
print(len(s) == len(set(s)))
```

```python
#✅ 4️⃣ Find smallest missing positive number 

arr = [3,4,-1,1]
arr = set(arr) 

i = 1
while i in arr:
    i += 1

print(i) 
```

```python
#✅ 5️⃣ Count frequency of numbers in list 

arr = [1,2,2,3,3,3]
freq = {x: arr.count(s) for x in set(arr)}
print(freq)
```
