```python
# ✅ 1. String Basics

s = "Divyanshu"
print(len(s))
print(s.lower())
print(s.upper())
print(s[::-1])  #reverse
```

```python
# ✅ 2. String Slicing
s = "Python" 
print(s[0:3]) #pyt
print(s[2:5])
print(s[-3:]) 
```

```python
# ✅ 3. Check substring

s = "python developer" 
print("python" in s)
```

```python
# ✅ 4. Counting characters

s = "banana" 
print(s.count("a")) 
```

```python
# ✅ 5. File Handling (MOST IMPORTANT)

#Reading File

f = open("data.txt","r") 
print(f.read())
f.close() 
```

```python
#Writing a file

f = open("data.txt","w") 
f.write("Hello world")
f.close() 
```

```python
#Appending 
f = open("data.txt","a") 
f.write("\n line added") 
f.close() 
```
