✅ DAY-5 INTERVIEW QUESTION

Q: What is the difference between "w" and "a" modes in file handling?

✅ Answer:

Mode	Meaning	Behavior
w	write	Overwrites file every time
a	append	Adds new data without deleting old

Example:

open("file.txt", "w") → replaces content  
open("file.txt", "a") → appends content
