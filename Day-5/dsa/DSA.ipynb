```python
# Reverse each word in a string 

s = "hello world"
words = s.split()
rev = [w[::-1] for w in words]
print(" ".join(rev)) 
```

```python
# 2️⃣ Check if array contains duplicate

arr = [1,2,3,1]
print(len(arr) != len(set(arr)))
```

```python
# 3️⃣ Find index of first occurrence

arr = [10,20,30,20]
target = 20
print(arr.index(target)) 
```

```python
# 4️⃣ Find all pairs whose sum = target
arr = [1,2,3,4,5]
target = 6

for i in range(len(arr)):
    for j in range(i+1, len(arr)):
        if arr[j] + arr[i] == target:
            print(arr[i], arr[j])
```

```python
# 5️⃣ Count characters in string using dictionary
s = "banana" 
freq = {} 
for ch in s :
    freq[ch] = freq.get(ch, 0) + 1
print(freq)
```
