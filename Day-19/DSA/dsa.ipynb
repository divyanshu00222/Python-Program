```python
#1️⃣ Find the Kth smallest element 
def kth_smallest(arr, k):
    arr.sort()
    return arr[k - 1]

print(kth_smallest([7, 10, 4, 3, 20, 15], 3))  # Output: 7

```

```python
#2️⃣ Check if two arrays are equal after sorting 

def are_equal(a, b):
    return sorted(a) == sorted(b)

print(are_equal([1, 2, 3], [3, 2, 1]))  # True
```

```python
#3️⃣ Merge two sorted arrays 

def merge_sorted(a, b):
    return sorted(a + b)

print(merge_sorted([1, 3, 5], [2, 4, 6]))
```

```python
#4️⃣ Sort array of 0s, 1s, and 2s 

def sort012(arr):
    low, mid, high = 0, 0, len(arr) - 1
    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1; mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1
    return arr

print(sort012([0, 1, 2, 0, 1, 2, 1, 0]))
```

```python
#5️⃣ Count Inversions in Array (Using Merge Sort) 
def merge_sort_count(arr):
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr)//2
    left, inv_left = merge_sort_count(arr[:mid])
    right, inv_right = merge_sort_count(arr[mid:])

    merged, inv_split = merge_and_count(left, right)
    return merged, inv_left + inv_right + inv_split

def merge_and_count(left, right):
    i = j = inv_count = 0
    merged = []
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
            inv_count += len(left) - i
    merged += left[i:]
    merged += right[j:]
    return merged, inv_count

arr = [1, 20, 6, 4, 5]
_, inv = merge_sort_count(arr)
print("Inversions:", inv)  # Output: 5

```
