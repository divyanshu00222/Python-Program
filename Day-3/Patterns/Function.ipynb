```python
#What is functions
def greet():
    print("Hello!")

greet()
```

```python
#Functions With Parameters

def add(a,b):
    return a + b

print(add(3, 5))
```

```python
#Default Parameters
def Welcome(name = "User"):
    print("Welcome", name)

Welcome()
Welcome("Divyanshu")
```

```python
#Return Statement
def square(n):
    return n * n

result = square(5)
print(result)
```

```python
#Pattern Basics(important for logic)
#Right Triangle

for i in range(1,5):
    print("*" * i)
    
```

```python
#Inverted Triangle 
for i in range(4,0,-1):
    print("*" * i)
    
```

```python
rows = 4
for i in range(rows):
    print(" "*(rows-i-1) + "*"*(2*i+1))
```

```python
for i in range(1, 5):
    for j in range(1, i+1):
        print(j, end = "")
    print()
```
