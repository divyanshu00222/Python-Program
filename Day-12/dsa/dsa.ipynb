```python
#✅ 1️⃣ Find factorial recursively 

def fact(n):
    return 1 if n == 0 else n * fact(n-1) 
print(fact(6)) 
```

```python
#✅ 2️⃣ Sum of list recursively 

def sum_list(lst):
    if not lst:
        return 0 
    return lst[0] + sum_list(lst[1:])

print(sum_list([1,2,3,4])) 
```

```python
#✅ 3️⃣ Find power (x^n) 

def power(x,n):
    if n == 0:
        return 1 
    return x * power(x, n-1)

print(power(2,5)) 
```

```python
#✅ 4️⃣ Count occurrences of element recursively 
def count_occ(lst, x):
    if not lst:
        return 0 
    
    return(lst[0] == x) + count_occ(lst[1:],x)

print(count_occ([1,2,3,1,1,4],1))
```

```python
#✅ 5️⃣ Reverse string recursively 
def reverse(s):
    if len(s) == 0:
        return s
    return reverse(s[1:]) + s[0]

print(reverse("python"))
```
