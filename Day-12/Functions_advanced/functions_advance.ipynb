```python
#✅ Default & Keyword Arguments 

def greet (name = "User"):
    print(f"Hello, {name}!") 

greet()
greet("Divyanshu") 
```

```python
#✅ *args (Variable arguments) 

def add_all(*nums):
    return sum(nums) 

print(add_all(1,2,3,4)) 
```

```python
#✅ **kwargs (Keyword arguments) 

def show_details(**data):
    for k,v in data.items():
        print(f"{k} = {v}") 

show_details(name="Divyanshu",age=23,role="Developer") 
```

```python
#✅ Nested Function Example 

def outer():
    def inner():
        print("Inner function") 

    inner()

outer() 
```

```python
#✅ Function Returning Another Function (Closure) 

def multiply_by(n):
    def multiply(x):
        return x * n 
    return multiply 

double = multiply_by(2) 
print(double(5))  #10 
```
