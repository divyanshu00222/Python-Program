```python
#✅ Factorial 

def fact(n):
    if n == 0:
        return 1 
    return n * fact(n-1) 

print(fact(5)) 
```

```python
#✅ Fibonacci 

def fib(n):
    if n <= 1:
        return n 
    return fib(n-1) + fib(n-2) 

for i in range(6):
    print(fib(i)) 
```

```python
#✅ Sum of digits 

def sum_digits(n):
    if n == 0:
        return 0 
    
    return n % 10 + sum_digits(n // 10)

print(sum_digits(1234)) 
```
