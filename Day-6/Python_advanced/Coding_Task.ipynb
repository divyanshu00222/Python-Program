```python
#1️⃣ Write a function to calculate factorial using loop 

def factorial(n):
    fact = 1
    for i in range(1 , n+1):
        fact *= i
    return fact 

print(factorial(5)) 
```

```python
#2️⃣ Count vowels using map 

s = "divyanshu" 
vowels = "aeiouAEIOU"

count = len(list(filter(lambda x: x in vowels, s)))
print(count) 
```

```python
#5️⃣ Use reduce to find product of list 

from functools import reduce 
nums = [1,2,3,4]
product = reduce(lambda a,b: a*b,nums)
print(product) 
```
