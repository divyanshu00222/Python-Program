```python
#1️⃣ Print all subarrays 

arr = [1,2,3]
for i in range(len(arr)):
    for j in range(i, len(arr)):
        print(arr[i:j+1]) 
        
```

```python
#2️⃣ Check if string is rotation of another 

s1 = "abcde" 
s2 = "cdeab" 

print(s2 in (s1+s1)) 
```

```python
#3️⃣ Find majority element (> n/2 times) 

arr = [2,2,1,2,3,2,2]
print(max(set(arr), key= arr.count)) 
```

```python
#4️⃣ Move negative numbers to left 

arr = [1,-2,3,-4,5]
result = [x for x in arr if x < 0] + [x for x in arr if x >= 0]
print(result)
```

```python
#5️⃣ Two sum using hash map 

arr = [2,7,11,15]
target = 0 

map = {}
for i, num in enumerate(arr):
    diff = target - num 
    if diff in map:
        print(map[diff], i)
        break 
    map[num] = i 
```
