```python
#✅ 1. Default and Keyword Arguments

def greet(name = "User", msg = "Welcome"):
    print(f"{msg},{name}!") 

greet()
greet("Divyanshu","Hello") 
```

```python
#✅ 2. Lambda Functions (very important)

square = lambda x: x * x 
print(square(5)) 
```

```python
#✅ 3. map() 
# Apply a function to each element 

nums = [1,2,3,4]
res = list(map(lambda x: x * x, nums))
print(res) 
```

```python
#✅ 4. filter()
# Filter items from list 

nums = [1,2,3,4,5]
evens = list(filter(lambda x: x%2==0, nums)) 
print(evens) 
```

```python
#✅ 5. reduce() 

from functools import reduce 

nums = [1,2,3,4]
total = reduce(lambda a,b: a+b , nums)
print(total) 
```
