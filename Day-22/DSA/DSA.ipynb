```python
#1️⃣ Two-Sum Problem

#✔ Hash map solution (O(n))

def two_sum(nums, target):
    h = {}
    for i, num in enumerate(nums):
        diff = target - num
        if diff in h:
            return [h[diff], i]
        h[num] = i

```

```python
#2️⃣ Find First Non-Repeating Character 
def first_unique(s):
    freq = {}
    for c in s:
        freq[c] = freq.get(c, 0) + 1
    for i, c in enumerate(s):
        if freq[c] == 1:
            return i
    return -1
```

```python
#3️⃣ Check if Two Strings Are Anagrams 
def is_anagram(a, b):
    return sorted(a) == sorted(b)
```

```python
#Or using hashing: 

def is_anagram(a, b):
    if len(a) != len(b):
        return False
    freq = {}
    for c in a:
        freq[c] = freq.get(c, 0) + 1
    for c in b:
        if c not in freq:
            return False
        freq[c] -= 1
        if freq[c] < 0:
            return False
    return True
```

```python
#4️⃣ Find Duplicate Number in Array 

def find_duplicate(nums):
    seen = set()
    for n in nums:
        if n in seen:
            return n
        seen.add(n)
```

```python
#5️⃣ Group Anagrams 

from collections import defaultdict

def group_anagrams(words):
    h = defaultdict(list)
    for w in words:
        key = tuple(sorted(w))
        h[key].append(w)
    return list(h.values())
```
