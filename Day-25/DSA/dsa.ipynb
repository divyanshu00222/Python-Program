```python
#Q1. Count nodes in linked list \
def count_nodes(head):
    count = 0
    cur = head
    while cur:
        count += 1
        cur = cur.next
    return count

```

```python
#Q2. Find maximum element 

def find_max(head):
    maxi = float('-inf')
    cur = head
    while cur:
        maxi = max(maxi, cur.data)
        cur = cur.next
    return maxi
```

```python
#Q3. Sum of all nodes 

def sum_list(head):
    s = 0
    cur = head
    while cur:
        s += cur.data
        cur = cur.next
    return s
```
