A Linked List is a chain of nodes.
Each node stores:

value | next_pointer


Example
A â†’ B â†’ C â†’ None

```python
#ðŸ§± 2ï¸âƒ£ Node + LinkedList Class (Simple Code) 

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    # Insert at end
    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        cur = self.head
        while cur.next:
            cur = cur.next
        cur.next = new_node

    # Print list
    def display(self):
        cur = self.head
        while cur:
            print(cur.data, end=" -> ")
            cur = cur.next
        print("None")

    # Search
    def search(self, target):
        cur = self.head
        while cur:
            if cur.data == target:
                return True
            cur = cur.next
        return False

    # Delete a node
    def delete(self, target):
        if not self.head:
            return

        # If head is to be deleted
        if self.head.data == target:
            self.head = self.head.next
            return

        cur = self.head
        while cur.next:
            if cur.next.data == target:
                cur.next = cur.next.next
                return
            cur = cur.next
```

```python
#ðŸ§ª 3ï¸âƒ£ Example Usage 

ll = LinkedList()

ll.insert(10)
ll.insert(20)
ll.insert(30)

ll.display()           # 10 -> 20 -> 30 -> None
print(ll.search(20))   # True

ll.delete(20)
ll.display()           # 10 -> 30 -> None
```
