```python
# ✅ 1️⃣ Count vowels in a string 

s = "developer"
vowels = "AEIOUaeiou" 

count = sum(1 for ch in s if ch in vowels)
print(count) 
```

```python
# ✅ 2️⃣ Find duplicate elements in list 

arr = [1,2,3,2,4,1]
dup = set([x for x in arr if arr.count(x) > 1]) 
print(dup) 
```

```python
#✅ 3️⃣ Find max occurring character 

s = "mississippi" 
ch = max(set(s), key=s.count) 
print(ch) 
```

```python
# ✅ 4️⃣ Check if array is sorted 

arr = [1,2,3,4]
print(arr == sorted(arr)) 
```

```python
# ✅ 5️⃣ Rotate array right by 1 

arr = [1,2,3,4]
arr = [arr[-1]] + arr[:-1]
print(arr) 
```
