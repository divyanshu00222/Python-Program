```python
# ü™ú What is Recursion?

# Recursion is when a function calls itself to solve smaller parts of a problem.

# ‚úÖ Two essential parts:

# Base Case ‚Üí When to stop recursion.

# Recursive Case ‚Üí When the function calls itself.
```

```python
#üßÆ Example 1: Factorial using Recursion 

def factorial(n):
    if n == 0 or n == 1:
        return 1  # Base case
    else:
        return n * factorial(n - 1)  # Recursive case

print(factorial(5))  # Output: 120
```

```python
#üîÅ Example 2: Fibonacci using Recursion 
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n - 1) + fib(n - 2)

print(fib(6))  # Output: 8

```

```python
#üß≠ Example 3: Sum of List Elements 

def list_sum(arr):
    if not arr:
        return 0
    return arr[0] + list_sum(arr[1:])

print(list_sum([1, 2, 3, 4]))  # Output: 10
```

```python
#üß∞ Example 4: Reverse String using Recursion 

def reverse_string(s):
    if len(s) == 0:
        return ""
    return s[-1] + reverse_string(s[:-1])

print(reverse_string("divyanshu"))  # Output: uhsnayvid
```

```python
#üß© Example 5: Power Function (x‚Åø) 

def power(x, n):
    if n == 0:
        return 1
    return x * power(x, n - 1)

print(power(2, 5))  # Output: 32
```
