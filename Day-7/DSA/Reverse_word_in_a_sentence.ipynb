```python
#✅ 1️⃣ Reverse words in a sentence 
s = "hello world python" 
print(" ".join(s.split()[::-1])) 
```

```python
#✅ 2️⃣ Frequency of characters 

s = "mississippi" 
freq = {ch: s.count(ch) for ch in set(s)} 
print(freq) 
```

```python
#✅ 3️⃣ Find missing number from 1–n 

arr = [1,2,4,5]
n = 5
missing = n*(n+1)//2 - sum(arr)
print(missing) 
```

```python
#✅ 4️⃣ Find second largest element 

arr = [10,20,4,45,99]
arr = list(set(arr)) 
arr.sort()
print(arr[-2]) 
```

```python
#✅ 5️⃣ Merge two sorted lists 

a = [1,3,5]
b = [2,4,6]
result = sorted(a+b)
print(result) 
```
