```python
#✅ 1. Advanced String Concepts 
#String Slicing

s = "Divyanshu"
print(s[0:4]) #Divy
print(s[-3:]) ##shu
```

```python
#✅ Reversing a string 
s = "python" 
print(s[::-1]) 
```

```python
#✅ Checking palindrome 

s = "madam" 
print(s == s[::-1]) 
```

```python
#✅ Useful string functions: 
s = "hello world" 
print(s.upper())
print(s.title())
print(s.replace("world","python")) 
```

```python
#✅ 2. List Comprehension (Very Important for Interviews) 

nums = [1,2,3,4] 
squares = [x*x for x in nums]
print(squares) 
```

```python
#✅ Even numbers only 
evens = [x for x in nums if x % 2 == 0]
```

```python
#✅ If-else inside list comprehension 

labels = ["Even" if x%2==0 else "Odd" for x in nums] 
```

```python
#✅ Nested loops in comprehension 

pairs = [(x,y) for x in [1,2] for y in [3,4]] 
print(pairs) 
```
